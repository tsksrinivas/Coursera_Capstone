### Read the wikipedia page into Pandas dataframe. There are two tables, pick the first one which is of interest for this lab.

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = requests.get(url).content
df_list = pd.read_html(html)
df2=df_list[0]
```

### Name the columns as per lab instructions.

```python
df2.columns = ['PostalCode', 'Borough', 'Neighborhood'] 
df2
```

```python
df2.shape
```

### Drop rows where a borough is Not assigned.

```python
df2.drop(df2[df2.Borough == 'Not assigned'].index, inplace = True)
```

### Combine rows with multiple Neighborhood values for same Postalcode, Borough into one row.

```python
df2['Neighborhood'] = df2[['PostalCode', 'Borough', 'Neighborhood']].groupby(['PostalCode', 'Borough'])['Neighborhood'].transform(lambda x: ', '.join(x))
df2[['PostalCode', 'Borough', 'Neighborhood']].drop_duplicates()
```

### If a Neighborhood is marked as Not assigned, then mark the neighborhood will same as the borough.

```python
df2.loc[(df2.Neighborhood == 'Not assigned'),'Neighborhood']=df2.Borough
df2.reset_index(drop=True, inplace = True)
```
